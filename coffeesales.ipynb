```python
plt.figure(figsize=(12,6))
plt.title("Total sales by coffee type")
sns.barplot(x = 'coffee_name' , y = 'money' , data = df  ,hue = 'Time_of_Day')
plt.show()
```

```python
monthly_earning = df.groupby(['Month_name' , 'Time_of_Day'])['money'].sum().reset_index()
week_day_earning = df.groupby(['Weekdaysort' , 'Time_of_Day'])['money'].sum().reset_index()
```

```python
plt.figure(figsize=(12,7))
plt.title('Total sales by month')
sns.barplot(x = 'Month_name' , y = 'money' , data = monthly_earning , hue = 'Time_of_Day')
plt.xlabel('Month')
plt.ylabel("Total_earning")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = 'Weekdaysort' , y= 'money', data = week_day_earning  , hue = 'Time_of_Day')
plt.title("week_days_earning")
plt.xlabel("Days")
plt.ylabel('Total_sales')
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.countplot(x = 'hour_of_day', hue = 'coffee_name' ,data = df)
plt.title("Coffee Preferences by Hour of the Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Coffees Sold")
plt.show()
```

```python
percentage_share = df.groupby('coffee_name')['money'].sum()
percentage_share
```

```python
explode = [0.1 if value == percentage_share.max() else 0 for value in percentage_share]
plt.figure(figsize=(12,6))
plt.pie(percentage_share , labels = percentage_share.index , autopct = '%1.1f%%' , explode = explode , shadow = True)
plt.title('coffee sales share by type')
plt.show()
```

```python
plt.figure(figsize = (8,4))
sns.lineplot(x = 'Month_name' , y = 'money' , data = monthly_earning , marker = 'o')
plt.title('Sales trends accross the months')
plt.xlabel('Month_Names')
plt.ylabel('Sales every month')
plt.grid()
plt.show()
```
